Emotions Indicators In general: 
- speech rate (the number of words spoken per minute). In conversational contexts, faster speech rate tends to increase the persuasiveness of the communicated message (Miller, Maruyama, Beaber, & Valone, 1976) and has been associated with positive speaker attributes such as competence. speech rate is more often observed in low arousal states
with predominantly negative valence (such as sadness)
- the absolute length of utterances: speakers who exhibit more voice
breaks during speech formation are generally perceived as being less
extraverted (Mallory & Miller, 1958) and less competent (Mohammadi
& Vinciarelli, 2015).
- amplitude of a soundwave: is captured by the sound intensity
level or sound pressure level, usually described as the amount of power
per unit area (such as watt per square meter or centimeter) and
commonly measured in decibels (dB).
- intensity: Voices with greater sound intensity, relative to baseline, are
typically perceived as being higher in dominance lower
sound intensity can indicate that a person is in a fearful, sad or tender
state
-Frequency: the number of cycles of a soundwave per second of vibrating air particles....human hearing typically ranges from 20 Hz to 20 kHz
- Pitch: higher pitch indicating anger, fear, and happiness, and lower
pitch indicative of sadness and tenderness
